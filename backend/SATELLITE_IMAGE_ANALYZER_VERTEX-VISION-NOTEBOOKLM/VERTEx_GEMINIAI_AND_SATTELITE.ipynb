```python
# prompt: #convert tif to jpg for all files in folders in satellite_images folder, save locally

import os
from PIL import Image
import glob

def convert_tif_to_jpg(tif_path, jpg_path):
  try:
    img = Image.open(tif_path)
    img.convert("RGB").save(jpg_path)
    print(f"Converted {tif_path} to {jpg_path}")
  except Exception as e:
    print(f"Error converting {tif_path}: {e}")


def process_folder(folder_path):
  tif_files = glob.glob(os.path.join(folder_path, '*.tif'))
  for tif_file in tif_files:
    jpg_file = os.path.splitext(tif_file)[0] + '.jpg'
    convert_tif_to_jpg(tif_file, jpg_file)


if __name__ == '__main__':
  satellite_images_path = '/content/satellite_images'  # Replace with your actual path
  for root, dirs, files in os.walk(satellite_images_path):
    for dir in dirs:
        folder_path = os.path.join(root, dir)
        process_folder(folder_path)
```

```python
import base64
import os
import csv
from google import genai
from google.genai import types
from PIL import Image
import glob

def generate_summary(image_path, date):
    """Generates a summary for a given image using the Gemini AI model."""
    client = genai.Client(api_key=os.environ.get("GEMINI_API_KEY"))

    files = [client.files.upload(file=image_path)]
    model = "gemini-2.0-flash-thinking-exp-01-21"

    # Modify the prompt to include the date
    modified_prompt = prompt + f"\n\nDate of the image Capture: {date}"

    contents = [
        types.Content(
            role="user",
            parts=[
                types.Part.from_uri(file_uri=files[0].uri, mime_type=files[0].mime_type),
                types.Part.from_text(text=modified_prompt),
            ],
        )
    ]

    generate_content_config = types.GenerateContentConfig(
        temperature=0.7,
        top_p=0.95,
        top_k=64,
        max_output_tokens=65536,
        response_mime_type="text/plain",
    )

    summary = ""
    for chunk in client.models.generate_content_stream(
        model=model, contents=contents, config=generate_content_config
    ):
        summary += chunk.text

    return summary


def process_images_and_create_csv(satellite_images_path, output_csv_file):
    """Processes images, generates summaries, and writes them to a CSV file."""
    with open(output_csv_file, "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Date", "Summary"])  # Write header row

        for root, dirs, files in os.walk(satellite_images_path):
            for dir in dirs:  # Iterate through folders (dates)
                date = dir  # Folder name represents the date
                folder_path = os.path.join(root, dir)
                image_files = glob.glob(os.path.join(folder_path, "*.jpg"))  # Get all JPGs

                for image_file in image_files:
                    try:
                        summary = generate_summary(image_file, date)
                        writer.writerow([date, summary])
                        print(f"Processed: {image_file}, Date: {date}")

                    except Exception as e:
                        print(f"Error processing {image_file}: {e}")


if __name__ == "__main__":
    satellite_images_path = "/content/satellite_images"
    output_csv_file = "satellite_image_summaries.csv"
    process_images_and_create_csv(satellite_images_path, output_csv_file)
```
